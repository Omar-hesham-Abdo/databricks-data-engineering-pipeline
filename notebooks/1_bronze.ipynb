```python
import requests
import json
import pandas as pd
from pyspark.sql.functions import current_timestamp,lit
import pyspark.sql.functions as F
catalog_name='products'
```

```python
def load_api():
    url='https://dummyjson.com/products'
    response=requests.get(url)
    data=response.json()
    return data['products']
products=load_api()
```

```python


pdf=pd.json_normalize(products)
pdf["price"] = pdf["price"].astype("float")
pdf["rating"] = pdf["rating"].astype("float")
pdf["stock"] = pdf["stock"].astype("int")
pdf["id"] = pdf["id"].astype("int")
pdf['discountPercentage']=(pdf['discountPercentage'].astype('float'))

pdf['weight']=(pdf['weight'].fillna(pdf['weight'].median()).astype('int'))

pdf['minimumOrderQuantity']=\
(pdf['minimumOrderQuantity'].fillna(pdf['minimumOrderQuantity'].median()).astype('int'))
```

```python
df_bronze=spark.createDataFrame(pdf)
df_bronze=df_bronze.withColumn('ingested_at',F.current_timestamp())
df_bronze=df_bronze.withColumn('data_source',lit('dummyjson_api'))

```

```python
df_bronze.write\
    .format('delta')\
        .mode('overwrite')\
            .option('overwriteSchema','true')\
                .saveAsTable(f'{catalog_name}.bronze.brz_products')
```
